## Problem 5: The Penguins Dataset

```python
url = "https://raw.githubusercontent.com/mcnakhaee/palmerpenguins/refs/heads/master/palmerpenguins/data/penguins.csv"
df = pd.read_csv(url)

df.describe()
```

### Analysis of Descriptive Statistics
The dataset was loaded from a raw Githib URL into pandas DataFrame. It contains measurements for different penguin species, including bill length, bill depth, flipper length, and body mass.

The mean represents the average value for each variable. It provides a general indication of the typical size of penguins in the dataset.

The standard deviation (std) measures how spread out the values are from the mean. A higher standard deviation indicates greater variability in the measurements, while a lower value suggests the data points are closer to the average.

The quartiles divide the dataset into four equal parts. The 25th percentile (Q1) shows the lower range of the data, the 50th percentile (median) represents the middle value, and the 75th percentile (Q3) shows the upper range. Theses values help to understand the distribution and identify any potential skewness in the data.

Overall, these descriptive statistics provide  a summary of the penguin measurements and help to understand the central tendancy and variability within the dataset

## Problem 6: Visualising Distributions

```python
plt.hist(df["body_mass_g"].dropna())
plt.title("Distribution of Penguin Body Mass")
plt.xlabel("Body Mass (g)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.boxplot(df["body_mass_g"].dropna())
plt.title("Boxplot of Penguin Body Mass")
plt.show()
```

The Histogram shows the overall distributon of penguin body mass.It highlights the frequency of values within different ranges, allowing  us to see where most penguins fall in terms of weight. From the histogram , we can observe that most body mass values are concentrated roughly between 3500g and 4500g, with fewer penguins at the lower and higher extremes.

The boxplot provides a more compact summary of the data using key statistical measures. It clearly shows the median  body mass (around 4000g), as well as the spread of the data through teh interquartile range. The boxplot also highlights the minimum and maximum values and make it easier to identify any potential outliers.

While the histogram gives a detailed view of the distribution shape, the boxplot reveals the tendancy and spread more clearly. In particular, the boxplot shows the median and quartiles explicitly, which are not directly visible in the histogram. It also provides a clearer indication of the range and whether any extreme values (outliers) are present.

## Problem 7: Investigating Relationship

```python
species_list = df["species"].dropna().unique()

for species in species_list:
    subset = df[df["species"] == species]
    plt.scatter(subset["body_mass_g"], subset["bill_length_mm"], label=species)

plt.xlabel("Body Mass (g)")
plt.ylabel("Bill Length (mm)")
plt.title("Relationship between Body Mass and Bill Length")
plt.legend()
plt.show()
```

The scatterplot shows the relationship between body mass and bill lenght for different penguin species. ..
