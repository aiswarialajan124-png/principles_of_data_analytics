The boxplot provides a more compact summary of the data using key statistical measures. While the histogram gives a detailed view of the distribution shape, the boxplot reveals the tendancy and spread more clearly. ## Problem 7: Investigating Relationship

### Scatterplot by Species

```python
# Loop through species and plot seperately
for species in df["species"].dropna().unique():
    subset = df[df["species"] == species]
    plt.scatter(subset["body_mass_g"], subset["bill_length_mm"], label=species)

plt.xlabel("Body Mass (g)")
plt.ylabel("Bill Length (mm)")
plt.title("Body Mass vs Bill Length by Species")
plt.legend()
plt.show()
```

### Analysis of Scatterplot
The scatterplot shows the relationship between body mass and bill length for different penguin species.
 - Adelie
 - Gentoo
 - Chinstrap

The larger penguinss tend to have longer bills. As body mass increases, bill length increases as well.

Each penguin species is colour-coded, due to this distict groupings can be observed. Each species forms its own cluster, indicating that the relationship between body mass and bill length varies between species. For example, Gentoo penguins have higher body mass and longer bills compared to others. This suggests that while there is an overall positive trend, species differences play an important role in the relationship between  variables.

## Problem 8: Modelling the Data

```python
# Select Variables
x = df["body_mass_g"]
y = df["bill_length_mm"]

# Remove missing values
mask = x.notna() & y.notna()
x = x[mask]
y = y[mask]
```

```python
# Fit straight line (degree 1 polynomial)
coeffs = np.polyfit(x, y,1)

# Create the line function
line = np.poly1d(coeffs)
```

```python
# Scatter plot colored by species
for species in df["species"].dropna().unique():
    subset = df[df["species"] == species]
    plt.scatter(subset["body_mass_g"], subset["bill_length_mm"], label=species)

# Plot best-fit line
plt.plot(x, line(x), color='black', linewidth=2, label='Best Fit Line')

plt.xlabel("Body Mass (g)")
plt.ylabel("Bill Length (mm)")
plt.title("Linear Relationship between Body Mass and Bill Length")
plt.legend()
plt.show()
```

```python
# Predicted values
y_red = line(x)

# Calculate R-squared
ss_res = np.sum((y - y_red) ** 2)
ss_tot = np.sum((y - np.mean(y)) ** 2)

r_squared = 1 - (ss_res / ss_tot)

r_squared
```

### Interpretation of Result
A linear model was fitted to the relationship between body mass and bill length using numpy.ployfit. The resulting line of best fit is shown on teh scatter plot.

The R-squared value measures how well the model explains the variation in the data. It ranges from 0 to 1, where values closer to 1 indicate a stronger relationship.

The calculated R-squared value is approximately 0.35. This indicates  a moderate positive relationship between body mass adn bill length. As body mass increases, bill length tends to increase as well, but the relationship is not perfectly linear.

From the plot, it is also clear that different penguin species form distict clusters. This suggests that species has a significant influence on bill length, meaning the body mass alone is not fully accurate predictor.

Overall, while body mass provides some predictive value, other factors such as species differences must be considered to better explain variation in bill length.
